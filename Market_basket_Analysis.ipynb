```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('D:/Machine_learning/unsupervised/Associate_Rule_Mining/Market_Basket_Optimisation.csv', header=None)
```

```python
df.head()
```

```python
transactions = []
for i in range(0, len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0, len(df.columns))])
```

```python
from apyori import apriori
```

```python
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 2, min_length = 2, max_length = 2)
```

```python
results = list(rules)
```

```python
type(results)
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
